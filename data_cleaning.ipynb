# imports + read data

```python
import pandas as pd
import matplotlib.pyplot as plt
import json

PATH = "datasets/employee-churn/churn.csv"
dataset = pd.read_csv(PATH)
```

# data exploration

```python
print(f"the dataset has {dataset.shape[0]} rows and {dataset.shape[1]} columns")
print("the colums are \n", json.dumps(dict(enumerate(dataset.columns)), indent=4))
print(dataset.head())
```

```python
def bar_plot(df, column):
    """This function plots a bar plot and give information on the column passed as argument."""
    df = df.copy()

    print(f"In the column {column} there are {df[column].isna().sum()} missing values")
    print(f"The type of the column {column} is {df[column].dtype}")

    v = df[column].value_counts(dropna=False)
    print(v)
    v.plot.bar(title=column)
    plt.show()
```

```python
def plot_all_columns(df):
    columns = df.columns.to_list()

    columns.remove('enrollee_id')

    for c in columns:
        print(c)
        bar_plot(df, c)
        print('\n')
```

```python
plot_all_columns(dataset)
```
